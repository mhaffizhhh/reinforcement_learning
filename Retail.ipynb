Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import Dataset

```python
df = pd.read_csv('data.csv')
df
```

UCB Algorithm

```python
import math
N = len(df)
d = 8
retail_selected = []
sums_reward = [0] * d
nums_of_selection = [0] * d
total_reward = 0

for n in range(0, N):
  retail = 0
  max_upper_bound = 0
  for i in range(0, d):
    if nums_of_selection[i] > 0:
      average_reward = sums_reward[i] / nums_of_selection[i]
      delta = math.sqrt(3/2 * math.log(n + 1) / nums_of_selection[i])
      upper_bound = average_reward + delta
    else:
      upper_bound = 1e400
    if upper_bound > max_upper_bound:
      max_upper_bound = upper_bound
      retail = i
  retail_selected.append(retail)
  reward = df.values[n, retail]
  sums_reward[retail] = sums_reward[retail] + reward
  nums_of_selection[retail] = nums_of_selection[retail] + 1
  total_reward = total_reward + reward
```

Plot the results

```python
plt.hist(retail_selected)
plt.xlabel('Retail')
plt.ylabel('Number of selection')
plt.show()
```
